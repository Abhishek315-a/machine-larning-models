# Hierarchical clustering
Hierarchical clustering is an unsupervised learning technique that builds a hierarchy (tree) of clusters. It doesnâ€™t need the number of clusters upfront and can be visualized with dendrograms. There are two types:

# Agglomerative (bottom-up):
Each point starts in its own cluster, clusters merge step-by-step.

# Divisive (top-down):
Starts with all points in one cluster and splits recursively.

# How Agglomerative Hierarchical Clustering Works
Begin with each data point as its own cluster.

Find the two closest clusters and merge them.

Recalculate distances between clusters.

Repeat steps 2-3 until all points form one cluster or stopping criteria are met.

# import necessary library

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

# Step 2: Load the data

```python
dataset = load_iris()
X = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

# Step 3: Scale the features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Step 4: Create and plot dendrogram to decide number of clusters

```python
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(X_scaled, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Sample index")
plt.ylabel("Euclidean distances")
plt.show()
```

# Step 5: Apply Agglomerative Clustering (choose clusters based on dendrogram)

```python
agglomodel = AgglomerativeClustering(n_clusters=3,metric="euclidean",linkage="ward")
labels=agglomodel.fit_predict(X_scaled)
```

# Step 6: Visualize clustering result (using first 2 features)

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='rainbow')
plt.title("Agglomerative Hierarchical Clustering - Iris")
plt.xlabel("Sepal length (scaled)")
plt.ylabel("Sepal width (scaled)")
plt.show()
```
