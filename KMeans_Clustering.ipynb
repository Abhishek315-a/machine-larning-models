# K-Means
K-Means is an unsupervised clustering algorithm that divides data points into
k
k groups (clusters) so that points in the same group are more similar to each other than to those in other groups. The algorithm iteratively assigns points to clusters and updates cluster centers (centroids).

# K-Means Algorithm Overview
Choose the number of clusters
k.

Randomly initialize
k
centroids.

Assign each data point to the nearest centroid (cluster assignment).

Update centroids by calculating the mean position of all points in each cluster.

Repeat steps 3 and 4 until centroids do not change significantly (convergence).



# Step 1: Import required libraries

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

# Step 2: Load the example dataset (Iris, drop labels for clustering)

```python
dataset = load_iris()
dataset
X = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

# Step 3: Optional - Normalize features for better clustering

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Step 4: Initialize and fit KMeans for k=3 clusters (since Iris has 3 species)

```python
Kmeansmodel = KMeans(n_clusters=3,random_state=42)
Kmeansmodel.fit(X_scaled)
```

# Step 5: Review results

```python
clusters = Kmeansmodel.labels_
centroid = Kmeansmodel.cluster_centers_
```

```python
print("Cluster assignments:", clusters)
print("Centroids:\n", centroid)
```

# Step 6: Visualize clusters using two features

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', marker='o')
plt.scatter(centroid[:, 0], centroid[:, 1], c='red', marker='x', s=200)
plt.xlabel("Sepal length (normalized)")
plt.ylabel("Sepal width (normalized)")
plt.title("K-Means clustering (Iris dataset)")
plt.show()
```
